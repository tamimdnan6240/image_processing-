```python
## define your rotate function for image

def rotate(image, angle):
    image = np.array(image) ## convert all images to numpy array
    (h,w) = image.shape[:2] ##  image.shape[:2], shape is a property of an image in Python that returns a tuple representing the dimensions of the imag
    center = (w//2, h//2)
    M = cv2.getRotationMatrix2D(center, angle*(-1), 1.0) ## 1 is the scaling factor to keep the same size
    image = cv2.warpAffine(image, M, (w,h))
    image = Image.fromarray(image)
    return image
```

```python
## rotate to 10 

# load all of the resized images
input_folder = "C:\\Users\\tamim\\OneDrive - University of North Carolina at Charlotte\\DeepCrack_(tamim)\\DeepCrack_Master\\codes\\data\\resized_mask"

# define the output folder
output_folder = 'C:\\Users\\tamim\\OneDrive - University of North Carolina at Charlotte\\DeepCrack_(tamim)\\DeepCrack_Master\\codes\\data\\rotated_mask'

# create output diretory if it doesn't exist
os.makedirs(output_folder, exist_ok = True)

## define angle range
angle = 10

# loop through each_image in the input directory 
for image_name in os.listdir(input_folder):
    # read the image
    image_path = os.path.join(input_folder, image_name)
    image = Image.open(image_path)
    # rotate the image
    image_rotated = rotate(image, angle)
    # save the rotated image to the output folder with a new filename
    output_name = f'resized_rotated_{angle}_{image_name}'
    output_path = os.path.join(output_folder, output_name)
    image_rotated.save(output_path)
```
