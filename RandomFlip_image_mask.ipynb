```python
## Call the function 
flp_img, flp_msk = RandomFlip(rot_30, mask_rot_30, FLIP_LEFT_RIGHT=True, FLIP_TOP_BOTTOM=True)

flp_img = cv2.cvtColor(flp_img, cv2.COLOR_RGB2BGR)
flp_msk = cv2.cvtColor(flp_msk, cv2.COLOR_RGB2BGR)

# make axes: images are arranged over one row
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

# you can change the size of the plot is required
fig_size = (12, 10)  # example base size

fig.set_figwidth(fig_size[0] * 2)  # where 2 is #columns
fig.set_figheight(fig_size[1] * 2)  # 2 is #rows

# plot the images on each axis
ax1.imshow(rot_30)      ## Original
ax1.imshow(flp_img)  ## new image

# additionally you can specify a colormap, e.g.
ax2.imshow(mask_rot_30, cmap='binary')
ax2.imshow(flp_msk, cmap='binary')  ## new image

# you can set the title, axis labels, legends, as usual, e.g.
ax1.set_title('Image')
ax2.set_title('Mask')

# removes borders
fig.tight_layout()

# finally show the images
plt.show()

## Saving the images


cv2.imwrite("rot_30_up_down_12.JPG", flp_img)
cv2.imwrite("rot_30_up_down_mask_12.png", flp_msk)
```
